````python
import pandas as pd
import matplotlib
import numpy as np
````

````python
genshin_df = pd.read_csv(r'C:\Users\Roast\Desktop\genshin_yt.csv', index_col=0)
genshin_df.head(5)
````

### Different Types of Videos

```
Version Trailers 
Special Programs
Character Teasers
Character Demos
Collected Miscellany
OSTs/Songs
```

````python
ver_trailer = genshin_df[genshin_df['Title'].str.contains(r'(?=.*Trailer)(?=.*Version)', 
                                        regex=True, case=False)]
ver_trailer
````

````python
spe_prog = genshin_df[genshin_df['Title'].str.contains('Program', regex=True, case=False)]
spe_prog
````

````python
tale  = genshin_df[genshin_df['Title'].str.contains(r'(?=.*Character)(?=.*Tale)', 
                                        regex=True, case=False)]
c_teaser = genshin_df[genshin_df['Title'].str.contains(r'(?=.*Character)(?=.*Teaser)', 
                                        regex=True, case=False)]
char_teaser = pd.concat([tale, c_teaser])
char_teaser
````

````python
c_demo = genshin_df[genshin_df['Title'].str.contains(r'(?=.*Character)(?=.*Demo)', 
                                        regex=True, case=False)]
c_chap = pd.DataFrame([genshin_df.iloc[216], genshin_df.iloc[220]])

char_demo = pd.concat([c_demo, c_chap])
char_demo
````

````python
coll_mis = genshin_df[genshin_df['Title'].str.contains(r'(?=.*Collected Miscellany)', 
                                        regex=True, case=False)]
coll_mis
````

````python
discs = genshin_df[genshin_df['Title'].str.contains(r'(?=.*disc)', regex=True, case=False)]
osts =  genshin_df[genshin_df['Title'].str.contains(r'(?=.*OST)', regex=True, case=True)]
eps = genshin_df[genshin_df['Title'].str.contains(r'(?=.*EP)', regex=True, case=True)]

songs = pd.concat([discs, osts, eps])
songs
````

````python
import openpyxl
import xlsxwriter
````

````python
all_df = [ver_trailer, spe_prog, char_demo, char_teaser, coll_mis, songs]

Excelwriter = pd.ExcelWriter("genshin_analysis.xlsx",engine="xlsxwriter")
for i, df in enumerate (all_df):
    df.to_excel(Excelwriter, sheet_name="Sheet" + str(i+1),index=False)
Excelwriter.save()
````
